# Решающие деревья (decision trees)

Дерево решений — это метод представления решающих правил в иерархической структуре, состоящей из элементов двух типов — узлов (node) и листьев (leaf). В узлах находятся решающие правила и производится проверка соответствия примеров этому правилу по какому-либо атрибуту обучающего множества.

## Принцип работы и визуализация


```python
# классификация
from sklearn.tree import DecisionTreeClassifier 
from sklearn import datasets
import pandas as pd
import numpy as np


iris = datasets.load_iris()
X = iris.data
y = iris.target

clf = DecisionTreeClassifier(max_depth=3)
clf.fit(X, y)
```

```python
from sklearn import tree

text_representation = tree.export_text(clf)
print(text_representation)
```

```python
from matplotlib import pyplot as plt

fig = plt.figure(figsize=(25,20))
_ = tree.plot_tree(clf, feature_names=iris.feature_names, class_names=iris.target_names, filled=True)
```

```python
# регрессия

from sklearn.tree import DecisionTreeRegressor 
from sklearn import datasets


diab = datasets.load_diabetes()
X = diab.data
y = diab.target

reg = DecisionTreeRegressor(max_depth=3)
reg.fit(X, y)
```

```python
text_representation = tree.export_text(reg)
print(text_representation)
```

```python
from matplotlib import pyplot as plt

fig = plt.figure(figsize=(25,20))
_ = tree.plot_tree(reg, feature_names=diab.feature_names, filled=True)
```

## Алгоритмы построения

**ID3** (Итеративный дихотомайзер 3) был разработан в 1986 году Россом Куинланом. 
Алгоритм создает дерево, находя для каждого узла (т. е. жадным способом) категориальный признак, который даст наибольший прирост информации. Деревья вырастают до максимального размера, а затем обычно применяется этап отсечения, чтобы улучшить способность дерева обобщать новые данные.

**C4.5** является преемником ID3 и убрал ограничение, согласно которому объекты должны быть категориальными, путем динамического определения дискретного атрибута (на основе числовых переменных), который разбивает непрерывное значение атрибута на дискретный набор интервалов. C4.5 преобразует обученные деревья (т. е. выходные данные алгоритма ID3) в наборы правил «если-то». Затем оценивается точность каждого правила, чтобы определить порядок их применения. Отсечение выполняется путем удаления предусловия правила, если точность правила улучшается без него.

**CART** (деревья классификации и регрессии) очень похож на C4.5, но отличается тем, что поддерживает числовые целевые переменные (регрессия) и не вычисляет наборы правил. CART строит бинарные деревья, используя признак и порог, которые дают наибольший прирост информации в каждом узле.

https://scikit-learn.org/stable/modules/tree.html#tree-algorithms-id3-c4-5-c5-0-and-cart

https://academy.yandex.ru/handbook/ml/article/reshayushchiye-derevya

Алгоритм CART
Строит бинарное дерево, где в узлах находится предикат, в листах находится ответ.
Задача -- минимизировать ошибку на каждом листе.

Алгоритм обучения упрощённо можно описать следующим образом:
0. Проверяем критерий остановки
1. Строим всевозможные разбиения на две подвыборки по одному признаку
2. Выбираем лучшее разбиение
3. Возвращаемся к шагу 0 для потомков
4. Проводим отсечение (pruning)

Описание алогритма: https://scikit-learn.org/stable/modules/tree.html#mathematical-formulation

```python
#решение
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn import metrics
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVR
from sklearn.svm import SVC
from sklearn.ensemble import BaggingRegressor
from sklearn.ensemble import BaggingClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import LinearSVC
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import StackingClassifier
from sklearn.linear_model import RidgeCV
from sklearn.svm import LinearSVR
from sklearn.ensemble import StackingRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
#подготовка датасета для регрессии
data_reg = pd.read_csv('../data/mumbai_houses_task.csv', sep=',')
null_dict = data_reg.isnull().sum()
type_dict = data_reg.dtypes
for key in null_dict.keys():
  if null_dict[key] != 0:
    if type_dict[key]=='int64':
      data_reg[key] = data_reg[key].fillna(data_reg[key].median())
    elif type_dict[key]=='float64':
      data_reg[key] = data_reg[key].fillna(data_reg[key].mean())
    else:
      data_reg[key] = data_reg[key].fillna(data_reg[key].mode())

data_reg = pd.get_dummies(data_reg)
data_reg['price'] = data_reg['price']/1000000
data_reg.rename(columns={'price':'price(mln)'}, inplace='True')

categorial_features = ["Status", "neworold", "Furnished_status", "type_of_building"]
real_features = ["area", "latitude", "longitude", "Bedrooms", "Bathrooms", "Balcony", "parking", "Lift"]
target_feature = "price(mln)"

learn_reg, test_reg = train_test_split(data_reg, test_size=0.2)
learn_reg_X = preprocessing.scale(learn_reg.drop(labels=target_feature, axis=1))
learn_reg_y = learn_reg[target_feature]
test_reg_X = preprocessing.scale(test_reg.drop(labels=target_feature, axis=1))
test_reg_y = test_reg[target_feature]
```

```python
#подготовка датасета для классификации
data_class = pd.read_csv('../data/smoke_detector_task.csv', sep=',')
data_class = data_class.drop(columns='Unnamed: 0')

null_dict = data_class.isnull().sum()
type_dict = data_class.dtypes
for key in null_dict.keys():
  if null_dict[key] != 0:
    if type_dict[key]=='int64':
      data_class[key] = data_class[key].fillna(data_class[key].median())
    elif type_dict[key]=='float64':
      data_class[key] = data_class[key].fillna(data_class[key].mean())
    else:
      data_class[key] = data_class[key].fillna(data_class[key].mode())

data_class = data_class.replace(['Yes', 'No'], [1, 0])

target_feature = 'Fire Alarm'

learn_class, test_class = train_test_split(data_class, test_size=0.2)
learn_class_X = learn_class.drop(labels=target_feature, axis=1).to_numpy()
learn_class_y = learn_class[target_feature].to_numpy()
test_class_X = test_class.drop(labels=target_feature, axis=1).to_numpy()
test_class_y = test_class[target_feature].to_numpy()

scaler = preprocessing.StandardScaler()
scaler.fit(learn_class_X)
learn_class_X = scaler.transform(learn_class_X)
test_class_X = scaler.transform(test_class_X)

data_class.to_csv("smoke_detector_task_edit.csv")

print(learn_class_X.shape)
print(test_class_X.shape)
```

```python
#метрики
def reg_metrics(y, pred_y):
    print('R2\t-\t', metrics.r2_score(y, pred_y))
    print('MSE\t-\t', metrics.mean_squared_error(y, pred_y))
    print('MAE\t-\t', metrics.mean_absolute_error(y, pred_y))
    print('MAPE\t-\t', metrics.mean_absolute_percentage_error(y, pred_y))

def class_metrics(y, y_pred):
    print('Accurancy \t - \t', metrics.accuracy_score(y, y_pred), '\n')
    print(metrics.classification_report(y, y_pred))
    print('ROC AUC \t - \t', metrics.roc_auc_score(y, y_pred))
```

```python
#1
#Регрессия
parameters = {'max_depth': np.arange(1,10,1)}
model_optimal = GridSearchCV(DecisionTreeRegressor(), parameters).fit(learn_reg_X, learn_reg_y)
max_depth_opt = model_optimal.best_params_['max_depth']
print(max_depth_opt)

regressor = DecisionTreeRegressor(max_depth=max_depth_opt)
regressor.fit(learn_reg_X, learn_reg_y)
y_reg_pred = regressor.predict(test_reg_X)
reg_metrics(y_reg_pred, test_reg_y)
```

```python
#Классификация
parameters = {'max_depth': np.arange(1,10,1)}
model_optimal = GridSearchCV(DecisionTreeClassifier(), parameters).fit(learn_class_X, learn_class_y)
max_depth_opt = model_optimal.best_params_['max_depth']
print(max_depth_opt)

clf = DecisionTreeClassifier(max_depth=max_depth_opt)
clf.fit(learn_class_X, learn_class_y)
y_class_pred = clf.predict(test_class_X)
class_metrics(y_class_pred, test_class_y)
```

```python
#2
parameters = {'n_estimators': np.arange(1,20,1)}
model_optimal = GridSearchCV(BaggingRegressor(), parameters).fit(learn_reg_X, learn_reg_y)
n_estimators_opt = model_optimal.best_params_['n_estimators']
print(n_estimators_opt)

breg = BaggingRegressor(n_estimators=n_estimators_opt)
breg.fit(learn_reg_X, learn_reg_y)
breg_pred = breg.predict(test_reg_X)
reg_metrics(breg_pred, test_reg_y)
```

```python
bclf = BaggingClassifier(n_estimators=10)
bclf.fit(learn_class_X, learn_class_y)
bclf_pred = bclf.predict(test_class_X)
class_metrics(bclf_pred, test_class_y)
```

```python
parameters = {'n_estimators': np.arange(1,5,1), 'max_depth': np.arange(2,5,1)}
model_optimal = GridSearchCV(GradientBoostingClassifier(), parameters).fit(learn_class_X, learn_class_y)
n_estimators_opt = model_optimal.best_params_['n_estimators']
max_depth_opt = model_optimal.best_params_['max_depth']
print(n_estimators_opt)
print(max_depth_opt)

gbclf = GradientBoostingClassifier(n_estimators=n_estimators_opt, learning_rate=1, max_depth=max_depth_opt)
gbclf.fit(learn_class_X, learn_class_y)
gbclf_pred = gbclf.predict(test_class_X)
class_metrics(gbclf_pred, test_class_y)
```

```python
gbreg = GradientBoostingRegressor(random_state=0)
gbreg.fit(learn_reg_X, learn_reg_y)
gbreg_pred = gbreg.predict(test_reg_X)
reg_metrics(gbreg_pred, test_reg_y)
```

```python
estimators = [
    ('rf', RandomForestClassifier(n_estimators=10, random_state=42)),
    ('svr', make_pipeline(StandardScaler(),
    LinearSVC(random_state=42)))
]
sclf = StackingClassifier(estimators=estimators, final_estimator=LogisticRegression())
sclf.fit(learn_class_X, learn_class_y)

sclf_pred = sclf.predict(test_class_X)
class_metrics(sclf_pred, test_class_y)
```

```python
estimators = [
    ('lr', RidgeCV()),
    ('svr', LinearSVR(random_state=42))
]
sreg = StackingRegressor(
    estimators=estimators,
    final_estimator=RandomForestRegressor(n_estimators=10, random_state=42)
)
sreg.fit(learn_reg_X, learn_reg_y)

sreg_pred = sreg.predict(test_reg_X)
reg_metrics(sreg_pred, test_reg_y)
```

```python
#4
#Реализация CART
class MyCART:
    tree = None
    
    def fit(self, X, y, max_depth=7):
        def gini(groups, classes):
            n_instances = sum([len(group) for group in groups])
            gini = 0
            for group in groups:
                size = float(len(group))
                if size == 0:
                    continue
                score = 0
                for class_val in classes:
                    p = [obj[-1] for obj in group].count(class_val) / size
                    score += p * p
                gini += (1.0 - score) * (size/n_instances)
            return gini
        
        def try_split(data, y, feature, value):
            left = []
            right = []
            for i in range(len(data)):
                if data[i][feature] < value:
                    left.append((data[i], y[i]))
                else:
                    right.append((data[i], y[i]))
            return left, right
        
        def get_split(X, y):
            class_values = list(set(y))
            b_feat, b_value, b_score, b_groups = 999,999,999,None
            for i_feat in range(len(X[0])):
                for obj in X:
                    groups = try_split(X, y, i_feat, obj[i_feat])
                    gini_index = gini(groups, class_values)
                    if gini_index < b_score:
                        b_feat, b_value, b_score, b_groups = i_feat, obj[i_feat],gini_index,groups
            return {'feat': b_feat, 'value': b_value, 'score': b_score, 'groups': b_groups}
        
        def mode(group):
            y = list(map(lambda x: x[1], group))
            return max(set(y), key=y.count)
        
        def split(node, max_depth, depth):
            left, right = node['groups']
            del(node['groups'])
            if not left or not right:
                node['left'] = node['right'] = mode(left + right)
                return
            if depth >= max_depth:
                node['left'], node['right'] = mode(left), mode(right)
                return

            node['left'] = get_split(list(map(lambda x: x[0], left)), list(map(lambda x: x[1], left)))
            split(node['left'], max_depth, depth+1)

            node['right'] = get_split(list(map(lambda x: x[0], right)), list(map(lambda x: x[1], right)))
            split(node['right'], max_depth, depth+1)
    
        
        self.tree = get_split(X, y)
        split(self.tree, max_depth, 1)
    
    def predict(self, X):
        def predict_rec(node, obj):
            if obj[node['feat']] < node['value']:
                if isinstance(node['left'], dict):
                    return predict_rec(node['left'], obj)
                else:
                    return node['left']
            else:
                if isinstance(node['right'], dict):
                    return predict_rec(node['right'], obj)
                else:
                    return node['right']
                
        pred = []
        for obj in X:
            pred.append(predict_rec(self.tree, obj))
        return pred
    
```

```python
model = MyCART()
model.fit(learn_class_X[:100], learn_class_y[:100])
pred = model.predict(test_class_X[:100])
class_metrics(test_class_y[:100], pred)
```
