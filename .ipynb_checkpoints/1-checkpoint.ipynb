<b>Предобработка</b>

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn import linear_model
from sklearn import metrics
from sklearn import preprocessing
from sklearn.model_selection import GridSearchCV

from sklearn.preprocessing import PolynomialFeatures
```

```python
data = pd.read_csv('../data/mumbai_houses_task.csv', sep=',')
null_dict = data.isnull().sum()
type_dict = data.dtypes
for key in null_dict.keys():
  if null_dict[key] != 0:
    if type_dict[key]=='int64':
      data[key] = data[key].fillna(data[key].median())
    elif type_dict[key]=='float64':
      data[key] = data[key].fillna(data[key].mean())
    else:
      data[key] = data[key].fillna(data[key].mode())

data
```

```python
data = pd.get_dummies(data)
data['price'] = data['price']/1000000
data.rename(columns={'price':'price(mln)'}, inplace='True')
data.to_csv('../data/mumbai_houses_task_edit.csv', index=False)
data
```

<b>Лабораторная работа</b>

Линейная регрессия

Без регуляризации:

```python
data = pd.read_csv('../data/mumbai_houses_task_edit.csv', sep=',')

categorial_features = ["Status", "neworold", "Furnished_status", "type_of_building"]
real_features = ["area", "latitude", "longitude", "Bedrooms", "Bathrooms", "Balcony", "parking", "Lift"]
target_feature = "price(mln)"
```

```python
fig, axes = plt.subplots(nrows=4, ncols=4, figsize=(15, 15))
for idx, feature in enumerate(data.columns[1:]):
    data.plot(feature, "price(mln)", subplots=True, kind="scatter", ax=axes[idx // 5, idx % 4])
```

```python
learn, test = train_test_split(data, test_size=0.2)
#learn_x = preprocessing.scale(learn[learn.columns[1:]])
#learn_x = np.hstack([preprocessing.scale(learn[real_features]), learn.drop(labels=[target_feature]+real_features, axis=1).to_numpy()])
learn_x = preprocessing.scale(learn.drop(labels=target_feature, axis=1))
learn_y = learn[target_feature]
test_x = preprocessing.scale(test.drop(labels=target_feature, axis=1))
test_y = test[target_feature]
model = linear_model.LinearRegression()
model.fit(learn_x, learn_y)
```

```python
test_predict = model.predict(test_x)
print('R2\t-\t', metrics.r2_score(test_y, test_predict))
print('MSE\t-\t', metrics.mean_squared_error(test_y, test_predict))
print('MAE\t-\t', metrics.mean_absolute_error(test_y, test_predict))
print('MAPE\t-\t', metrics.mean_absolute_percentage_error(test_y, test_predict))
#np.sqrt(((1/len(test))*(test[target_feature]- test_predict)**2).mean())
```

С регуляризацией:

```python
parameters = {'alpha': np.arange(0.2,50,0.2)}
model_optimal = GridSearchCV(linear_model.Lasso(), parameters).fit(learn_x, learn_y)
alpha_opt = model_optimal.best_params_['alpha']
print(alpha_opt)
```

```python
model_l1 = linear_model.Lasso(alpha=alpha_opt)
model_l1.fit(learn_x, learn_y)
test_l1_predict = model_l1.predict(test_x)
print('R2\t-\t', metrics.r2_score(test_y, test_predict))
print('MSE\t-\t', metrics.mean_squared_error(test_y, test_l1_predict))
print('MAE\t-\t', metrics.mean_absolute_error(test_y, test_l1_predict))
print('MAPE\t-\t', metrics.mean_absolute_percentage_error(test_y, test_l1_predict))
```

```python
parameters = {'alpha': np.arange(0.2,50,0.2)}
model_optimal = GridSearchCV(linear_model.Ridge(), parameters).fit(learn_x, learn_y)
alpha_opt = model_optimal.best_params_['alpha']
print(alpha_opt)
```

```python
model_l2 = linear_model.Ridge(alpha=alpha_opt)
model_l2.fit(learn_x, learn_y)
test_l2_predict = model_l2.predict(test_x)
print(model_l2.coef_)
print(model_l2.intercept_)
print('R2\t-\t', metrics.r2_score(test_y, test_l2_predict))
print('MSE\t-\t', metrics.mean_squared_error(test_y, test_l2_predict))
print('MAE\t-\t', metrics.mean_absolute_error(test_y, test_l2_predict))
print('MAPE\t-\t', metrics.mean_absolute_percentage_error(test_y, test_l2_predict))
#np.sqrt(((test[target_feature]- test_l2_predict)**2).mean())
```

```python
class MyRidge:
    def __init__(self, a):
        self.weights = 0
        self.intercept = 0
        self.a = a
    
    def grad(self, X, w, y):
        y_pred = X @ w
        return 2/len(X)*(y-y_pred) @ (-X) + self.a*2*np.sum(self.weights)
    
    def fit(self, X, y):
        X = np.column_stack([X, np.ones(len(X))])
        EPS = 0.0001
        next_weights = np.zeros(X.shape[1])
        LEARNING_RATE = 0.01
        n = 10000
        for i in range(n):
            self.weights = next_weights
            next_weights = self.weights - LEARNING_RATE*self.grad(X, self.weights, y)
            if np.linalg.norm(self.weights - next_weights, ord=2) <= EPS:
                break
        self.intercept = self.weights[-1]
        self.weights = self.weights[:-1]
        print(self.weights)
        print(self.intercept)

    def predict(self, X):
        return X @ self.weights + self.intercept
```

```python
def myR2(y1, y2):
    rmse2 = (1/len(y1) * np.sum((y1-y2)**2))
    y = 1/len(y1) * np.sum(y1)
    d = 1/len(y1) * np.sum((y1-y)**2)
    return 1 - rmse2 / d

def myMSE(y1, y2):
    return 1/len(y1)*np.sum((y1-y2)**2)

def myMAE(y1, y2):
    return 1/len(y1)*np.sum(np.absolute(y1-y2))

def myMAPE(y1, y2):
    return 1/len(y1)*np.sum(np.absolute(y1-y2)/np.absolute(y1))
```

```python
model = MyRidge(1)
model.fit(learn_x, learn_y)

test_predict = model.predict(test_x)
print('MyR2\t-\t', myR2(test_y, test_predict))
print('R2\t-\t', metrics.r2_score(test_y, test_predict))

print('MyMSE\t-\t', myMSE(test_y, test_predict))
print('MSE\t-\t', metrics.mean_squared_error(test_y, test_predict))

print('MyMAE\t-\t', myMAE(test_y, test_predict))
print('MAE\t-\t', metrics.mean_absolute_error(test_y, test_predict))

print('MyMAPE\t-\t', myMAPE(test_y, test_predict))
print('MAPE\t-\t', metrics.mean_absolute_percentage_error(test_y, test_predict))
```
