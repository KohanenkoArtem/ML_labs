```python
def kmeans_alg(data, data_y):
    # 
    parameters = {'n_clusters': np.arange(1, 3, 1),
              'n_init': np.arange(10,100,10),
              'max_iter': np.arange(100,1000,100)}
    kmeans_optimal = GridSearchCV(KMeans(), parameters).fit(data)
    print(':', kmeans_optimal.best_params_)
    #
    kmeans = KMeans(**kmeans_optimal.best_params_).fit(data)
    # 
    plt.scatter(data[:,0], data[:,1], c=kmeans.labels_)
    plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=100, c='red')
    #
    print("")
    print("Rand Score:", rand_score(kmeans.labels_, data_y))
    print("Adjusted Rand Score:", adjusted_rand_score(kmeans.labels_, data_y))
    print("")
    print("Silhouette Score:", silhouette_score(data, kmeans.labels_))
    print("CalinskiHarabasz Index:", calinski_harabasz_score(data, kmeans.labels_))
```

```python
kmeans_alg(data1, data1_y)
```

```python
kmeans_alg(data2, data2_y)
```

```python
kmeans_alg(data3, data3_y)
```

```python
def affinityprop_alg(data, data_y):
    # 
    parameters = {'damping': np.arange(0.6, 1.01, 0.1),
                  'preference': np.arange(-40,-1,1)}
    affinityprop_optimal = GridSearchCV(AffinityPropagation(), scoring = 'rand_score', param_grid = parameters).fit(data, data_y)
    print(':', affinityprop_optimal.best_params_)
    #
    affinityprop = AffinityPropagation(**affinityprop_optimal.best_params_).fit(data)
    #
    plt.scatter(data[:,0], data[:,1], c=affinityprop.labels_)
    plt.scatter(affinityprop.cluster_centers_[:,0], affinityprop.cluster_centers_[:,1], s=100, c='red')
    #
    print("")
    print("Rand Score:", rand_score(affinityprop.labels_, data_y))
    print("Adjusted Rand Score:", adjusted_rand_score(affinityprop.labels_, data_y))
    print("")
    print("Silhouette Score:", silhouette_score(data, affinityprop.labels_))
    print("CalinskiHarabasz Index:", calinski_harabasz_score(data, affinityprop.labels_))
```

```python
def dbscan_alg(data, data_y):
    # 
    nbrs = NearestNeighbors(n_neighbors=4).fit(data)
    distances, indices = nbrs.kneighbors(data)
    distances = np.mean(distances, axis=1)
    distances = np.sort(distances)
    x = np.arange(1, distances.size + 1 ,1)
    kneedle = KneeLocator(x, distances, curve = 'convex', direction = 'increasing')
    print(':', 'Eps:', distances[kneedle.elbow - 1])
    #Elbow Method 
    figure, axis = plt.subplots(2)
    figure.set_figheight(10)
    figure.set_figwidth(10)
    axis[0].plot(x, distances)
    axis[0].vlines(kneedle.elbow, plt.ylim()[0], plt.ylim()[1], linestyles='dashed')
    #
    dbscan = DBSCAN(eps = distances[kneedle.elbow - 1]).fit(data)
    #
    axis[1].scatter(data[:,0], data[:,1], c=dbscan.labels_)
    #
    print("")
    print("Rand Score:", rand_score(dbscan.labels_, data_y))
    print("Adjusted Rand Score:", adjusted_rand_score(dbscan.labels_, data_y))
    print("")
    print("Silhouette Score:", silhouette_score(data, dbscan.labels_))
    print("CalinskiHarabasz Index:", calinski_harabasz_score(data, dbscan.labels_))
```

```python
dbscan_alg(data1, data1_y)
```

```python
dbscan_alg(data2, data2_y)
```

```python
dbscan_alg(data3, data3_y)
```

```python
def gaussianmixture_alg(data, data_y):
    # 
    parameters = {'n_components': np.arange(1, 10, 1)}
    gaussianmixture_optimal = GridSearchCV(GaussianMixture(), parameters).fit(data)
    print(':', gaussianmixture_optimal.best_params_)
    #
    gaussianmixture = GaussianMixture(**gaussianmixture_optimal.best_params_).fit(data)
    gaussian_pred = gaussianmixture.predict(data)
    #
    plt.scatter(data[:,0], data[:,1], c=gaussian_pred)
    plt.scatter(gaussianmixture.means_[:,0], gaussianmixture.means_[:,1], s=100, c='red')
    #
    print("")
    print("Rand Score:", rand_score(gaussian_pred, data_y))
    print("Adjusted Rand Score:", adjusted_rand_score(gaussian_pred, data_y))
    print("")
    print("Silhouette Score:", silhouette_score(data, gaussian_pred))
    print("CalinskiHarabasz Index:", calinski_harabasz_score(data, gaussian_pred))
```

```python
def agglomerativecl_alg(data, data_y, n_clusters):
    #
    agglomerativecl = AgglomerativeClustering(n_clusters = n_clusters, linkage = 'ward').fit(data)
    #
    plt.scatter(data[:,0], data[:,1], c=agglomerativecl.labels_)
    #
    print("")
    print("Rand Score:", rand_score(agglomerativecl.labels_, data_y))
    print("Adjusted Rand Score:", adjusted_rand_score(agglomerativecl.labels_, data_y))
    print("")
    print("Silhouette Score:", silhouette_score(data, agglomerativecl.labels_))
    print("CalinskiHarabasz Index:", calinski_harabasz_score(data, agglomerativecl.labels_))

def create_dendogram(data):
    clusters = hierarchy.linkage(data, method="ward")
    plt.figure(figsize=(8, 6))
    dendrogram = hierarchy.dendrogram(clusters) 
    plt.axhline(25, color='red', linestyle='--'); 
    plt.axhline(15, color='crimson');
```

```python
class myKMeans:
    def __init__(self, n_clusters, iters):
        self.n_clusters = n_clusters
        self.iters = iters
        self.labels = None
        self.centers = None
        
    def kmeans_plus_plus(self, X, n_clusters):
        idx = random.randrange(len(X))
        centroids = [X[idx]]
        for _ in range(1, n_clusters):
            squared_distances = np.array([min([np.inner(centroid - sample,centroid - sample) for centroid in centroids]) for sample in X])
            proba = squared_distances / squared_distances.sum()
            for point, probability in enumerate(proba):
                if probability == proba.max():
                    centroid = point
                    break
  
            centroids.append(X[centroid])
  
        return np.array(centroids)
  
    def find_closest_centroids(self, X, centroids):
        K = centroids.shape[0]
        label = np.zeros(X.shape[0], dtype=int)
        for sample in range(len(X)):
            distance = []
            for centroid in range(len(centroids)):
                norm = np.linalg.norm(X[sample] - centroids[centroid])
                distance.append(norm)
            label[sample] = distance.index(min(distance))
        return label
  
    def compute_centroids(self, X, idx, K):
        m, n = X.shape
        centroids = np.zeros((K, n))
        for k in range(K):   
            points = X[idx == k]
            centroids[k] = np.mean(points, axis=0)
        return centroids
  
    def fit(self, X):
        m, n = X.shape
        initial_centroids = self.kmeans_plus_plus(X, self.n_clusters)
        centroids = initial_centroids   
        labels = np.zeros(m)
        prev_centroids = centroids
        for i in range(self.iters):
            labels = self.find_closest_centroids(X, centroids)
            centroids = self.compute_centroids(X, labels, self.n_clusters)
            if centroids.tolist() == prev_centroids.tolist():
                break
            else:
                prev_centroids = centroids
        self.labels = labels
        self.centers = centroids
```

```python
mykmeans = myKMeans(2, 50)
mykmeans.fit(data1)
```

```python
plt.scatter(data1[:, 0], data1[:, 1], c=mykmeans.labels) 
plt.scatter(mykmeans.centers[:, 0], mykmeans.centers[:, 1], s=100, c='red')
print("")
print("Rand Score:", rand_score(mykmeans.labels, data1_y))
print("Adjusted Rand Score:", adjusted_rand_score(mykmeans.labels, data1_y))
print("")
print("Silhouette Score:", silhouette_score(data1, mykmeans.labels))
print("CalinskiHarabasz Index:", calinski_harabasz_score(data1, mykmeans.labels))
```
